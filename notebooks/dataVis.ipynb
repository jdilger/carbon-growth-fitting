```python

import pandas as pd
import numpy as np
data_path = 'count_standage_by_standage_1990_2021_startage_80.csv'
reducer = 'count'
df = pd.read_csv(data_path)
df =df.loc[(df['readable'] != 'non forest') & (df['readable'] != 'forest loss')  & (df['readable'] != 'masked areas outside of aoi')]
df['year'] = df['map_name'].str.split('_').str[1]
df['sumHa'] = df[reducer].apply(lambda i : i * 30 * 30 / 10000)
df['order'] = df['readable'] 
orderDict = { 'stand age 40+':50,
             'stand age 30-40':40,
             'stand age 20-30':30,
             'stand age 10-20':20,
             'stand age 5-10':10,
             'stand age 1-5':1
            }
df.replace({'order':orderDict})
```

```python

import altair as alt
chart_title = "area of forest by stand age 1990-2021".title()

alt.Chart(df, title=chart_title).mark_bar(
    cornerRadiusTopLeft=3,
    cornerRadiusTopRight=3
).encode(
    alt.X('year:O', title='Years'),
    alt.Y('sumHa:Q', title='HA', axis=alt.Axis(format='~s')),#sort=alt.EncodingSortField(field='readable', order='ascending')
    alt.Color('readable:O',  legend=alt.Legend(title="Stand age by color"), #sort=alt.EncodingSortField(field='sumHa', order='descending')
              scale=alt.Scale(domain = [ 'stand age 40+','stand age 30-40', 'stand age 20-30','stand age 10-20','stand age 5-10','stand age 1-5'],
#                               scheme='greens'
                              range=['#69a84f','#47bdc6','#e69138','#f1c232','#b5a7d6','#ea4335']
                             ),
              # https://vega.github.io/vega/docs/schemes/#reference
             ),
    order='sumHa:O'
)
```

```python
data_path = 'sum_carbonPlantation_by_standage_1990_2021_startage_80.csv'
reducer = 'sum'
df = pd.read_csv(data_path)
df =df.loc[(df['readable'] != 'non forest') & (df['readable'] != 'forest loss')  & (df['readable'] != 'masked areas outside of aoi')]
df['year'] = df['map_name'].str.split('_').str[1]
df['sumHa'] = df[reducer].apply(lambda i : i * 30 * 30 / 10000)
```

```python
chart_title = "sum of forest carbon by stand age 1990-2021 (plantation)".title()

alt.Chart(df, title=chart_title).mark_bar(
    cornerRadiusTopLeft=3,
    cornerRadiusTopRight=3
).encode(
    alt.X('year:O', title='Years'),
    alt.Y('sumHa:Q', title='Carbon', 
          axis=alt.Axis(format='~s'),
          sort=alt.EncodingSortField(field='readable', order='ascending'),
         scale=alt.Scale(domain=[0,450000000]),
         ),
    alt.Color('readable:O',  
              sort=alt.EncodingSortField(field='sumHa', order='descending'),
              legend=alt.Legend(title="Stand age by color"), 
              scale=alt.Scale(domain = [ 'stand age 40+','stand age 30-40', 'stand age 20-30','stand age 10-20','stand age 5-10','stand age 1-5','stand age 40+'],
                              range=['#69a84f','#47bdc6','#e69138','#f1c232','#b5a7d6','#ea4335']
              ),
              # https://vega.github.io/vega/docs/schemes/#reference
             ),
#     scheme=['#69a84f','#47bdc6','#e69138','#f1c232','#b5a7d6','#ea4335'],
    order='sumHa'
)
```
